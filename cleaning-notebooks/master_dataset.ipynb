```python
# Not all are necessary
import sys
import os
#import numpy as np
import matplotlib.pyplot as plt
#import tensorflow as tf
import pandas as pd

from data_cleaning.data_cleaner import DataCleaner
```

```python
project_root = os.getcwd()
sys.path.append(os.path.abspath(project_root))
```

```python
dc: DataCleaner = DataCleaner(project_root + '/raw_data/Sea_level_1_raw.csv')
#dc.handle_missing(column='AverageTemperature', strategy='drop')
print(dc.count_missing_values())
dc.preview(10)
```

```python
# Drop unneeded columns
columns_to_drop = ['Forest Area (%)', 'Renewable Energy (%)', 'Rainfall (mm)', 'Population']
dc.drop(columns_to_drop)
dc.preview(10)
```

```python
# Group by Target Country
target_country = 'Russia' # Save file for different countries (only filtered by this specific country)
test_df = dc.df[dc.df['Country'] == target_country]
print(test_df['Year'].value_counts())
```

```python
# Group the data by 'Year' and calculate the required metrics
processed_data = test_df.groupby('Year').agg(
    avg_global_temp=('Avg Temperature (°C)', 'mean'),  # Calculate average global temperature
    total_co2_emissions=('CO2 Emissions (Tons/Capita)', 'mean'),  # Calculate aggregate CO2 emissions
    sea_level_rise=('Sea Level Rise (mm)', 'mean'), # Calculate mean sea level rise 
    average_extreme_events=('Extreme Weather Events', 'mean') # Calculate mean annual extreme weather events
).reset_index()
```

```python
output_file_path = os.path.join(project_root, f"clean_data/master_dataset_{target_country}.csv")
processed_data.to_csv(output_file_path, index=False)  # Write the processed data to a CSV file

# Display the processed data to verify the results
processed_data.head()  # Display the first few rows of the processed DataFrame
```
