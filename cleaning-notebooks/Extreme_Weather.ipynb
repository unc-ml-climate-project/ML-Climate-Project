# Import Required Libraries and Modules and Read CSV File
* **pandas**: used for data manipulation  
* **os**: used for accessing file system paths


```python
import pandas as pd
import os

# Read the CSV File
project_root = "/Users/aashvijain/UNC/SPRING25/COMP562/ML-Climate-Project"
file_path = os.path.join(project_root, "raw-data", "Extreme_weather_raw.csv")
data = pd.read_csv(file_path)
data.head()  # Display the first few rows to verify the data
```

# Process Data to calculate averages and aggregates

```python
processed_data = data.groupby('Year').agg({
    'Avg Temperature (°C)': 'mean',           # Average temperature per year
    'CO2 Emissions (Tons/Capita)': 'mean',    # Average CO2 emissions per year
    'Sea Level Rise (mm)': 'mean',            # Average sea level rise per year
    'Rainfall (mm)': 'mean',                  # Average rainfall per year
    'Population': 'sum',                      # Total population per year
    'Renewable Energy (%)': 'mean',           # Average renewable energy percentage
    'Extreme Weather Events': 'sum',          # Total extreme weather events per year
    'Forest Area (%)': 'mean'                 # Average forest area percentage
}).round(2)  # Round all values to 2 decimal places

# Reset the index to make 'Year' a column again
processed_data = processed_data.reset_index()
```

# Save to new CSV file

```python
output_file_path = os.path.join(project_root, "clean-data", "processed_extreme_weather.csv")
processed_data.to_csv(output_file_path, index=False)

# Display the processed data to verify the results
processed_data.head()
```
