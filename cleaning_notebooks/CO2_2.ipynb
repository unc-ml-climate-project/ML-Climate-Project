```python
import pandas as pd
import os 
```

```python
# read CSV
data_path = "/Users/laurenfeldman/COMP 562/ML-Climate-Project/raw-data/CO2_2_raw.csv"
df = pd.read_csv(data_path)

df.head()
```

# Aggregate Yearly CO2 Emissions 

```python
# Convert 'Date' to datetime and extract the year
df['Year'] = pd.to_datetime(df['Date']).dt.year

grouped_data = df.groupby(['Year']).agg(
    Average_Temperature=('Temperature', 'mean'),
    CO2_emissions=('CO2 Emissions', 'sum')
).reset_index()

# Save Processed Data to a New CSV File
output_file_path = os.path.join("/Users/laurenfeldman/COMP 562/ML-Climate-Project/clean-data", "C02_2_global_averages.csv")
grouped_data.to_csv(output_file_path, index=False)  

grouped_data.head()  
```

# Aggregate Yearly CO2 Emissions Per Country

```python
grouped_data = df.groupby(['Year', 'Country']).agg(
    Average_Temperature=('Temperature', 'mean'),
    CO2_emissions=('CO2 Emissions', 'sum')
).reset_index()

# Save Processed Data to a New CSV File
output_file_path = os.path.join("/Users/laurenfeldman/COMP 562/ML-Climate-Project/clean-data", "C02_2_country_averages.csv")
grouped_data.to_csv(output_file_path, index=False)  

grouped_data.head()  
```
